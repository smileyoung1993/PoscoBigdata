Histgram으로 diagnosis (양성,음성)과의 분포도 살펴보자

```python
def double_hist(feature):
    plt.hist(df_breast[df_breast['diagnosis']==0][feature],label = "악성",alpha = 0.5)
    plt.hist(df_breast[df_breast['diagnosis']==1][feature],label = "양성",alpha = 0.5)
    plt.legend()
```
