<h2>histogram으로 양성,음성 판단<h2>

```python
def hist_map(feature):
    grid = sns.FacetGrid(df_breast, col = "diagnosis")
    grid.map(plt.hist,feature)
```

```python
# mean radius 가 작을수록 유방암 양성의 확률이 높다
hist_map("mean radius")
```

```python
# mean perimeter가 작을수록 유방암 양성의 확률이 높다.
hist_map("mean perimeter")
```

```python
# mean area가 작을수록 유방암 양성의 확률이 높다.
# diagnosis = 0인경우 데이터가 왼쪽으로 치우침, 변환이 필요하다
hist_map("mean area")
```

```python
# worst radius 가 작을수록 유방암 양성의 확률이 높다
hist_map("worst radius")
```

```python
# worst perimeter 가 작을수록 유방암 양성의 확률이 높다
hist_map("worst perimeter")
```

```python
# worst area 가 작을수록 유방암 양성의 확률이 높다
# diagnosis = 0인경우 데이터가 왼쪽으로 치우침, 변환이 필요하다
hist_map("worst area")
```

```python
# 매끄러움의 평균은 암의 악성 양성의 판단의 기준으로는 힘들다
# 단, 양성의 경우의 분포가 더 많이 나타난다.
hist_map("mean smoothness")
```

<h2> 중첩 histogram <h2>

```python
select_feature = df_breast[["diagnosis","mean radius","mean perimeter","mean area","worst radius","worst perimeter","worst area","mean concavity","mean concave points","worst concavity","worst concave points"]]
select_feature
```

```python
plt.figure(figsize= (16,16))
sns.heatmap(select_feature.corr(),annot= True, fmt = ".2f",linewidths =.5,cmap ='Blues' )
```

<h3>악성,양성에 대한 diagnosis 와 상관도가 높은 변수들 간의 연관성을 중첩히스트를 통해 확인<h3>

```python
def super_hist_map(feature):
    plt.hist(df_breast[df_breast["diagnosis"]==0][feature],label = "악성",alpha = 0.5)
    plt.hist(df_breast[df_breast["diagnosis"]==1][feature],label = "양성",alpha = 0.5)
    plt.legend()
    
```

```python
# x_features = df_breast.drop("diagnosis",axis = 1)

# # xaxes = ['x1','x2','x3','x4']
# # yaxes = ['y1','y2','y3','y4']
# # titles = ['t1','t2','t3','t4'] 

# # f,a = plt.subplots(2,2)
# # a = a.ravel()
# # for idx,ax in enumerate(a):
# #     ax.hist(data[idx])
# #     ax.set_title(titles[idx])
# #     ax.set_xlabel(xaxes[idx])
# #     ax.set_ylabel(yaxes[idx])
# # plt.tight_layout()
# x_features
features = x_features.columns
features
f,a = plt.subplots(5,6)
a = a.ravel
for feature,plt in enumerate(a):
    plt.hist(df_breast[df_breast["diagnosis"]==0][feature],label = "악성",alpha = 0.5)
    plt.hist(df_breast[df_breast["diagnosis"]==1][feature],label = "양성",alpha = 0.5)
    plt.legend()
plt.tight_layout()
```

```python
fig, axs = plt.subplots(5, 6, sharex=True, sharey=True)
axs[0,0].hist(df_breast[df_breast["diagnosis"]==0][feature],label = "악성",alpha = 0.5)
axs[0,0].hist(df_breast[df_breast["diagnosis"]==1][feature],label = "양성",alpha = 0.5)
fig.legend()
plt.show()
```
