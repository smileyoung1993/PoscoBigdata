```python
# 로지스틱회귀분석을 적용해야하지않을까..(진단결과가 0:악성, 1:양성)이산형이므로

#다중회귀분석
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns 
import statsmodels.api as sm
from sklearn.preprocessing import scale,robust_scale,minmax_scale
#다중공정성 패키지
from statsmodels.stats.outliers_influence import variance_inflation_factor
#sklearn 선형회귀 모형 :변수선택법
from sklearn.linear_model import LinearRegression
#변수 선택법: 후진제거법
from sklearn.feature_selection import RFE
# 데이터 스케일
from sklearn.preprocessing import scale
```

```python
df_breast_cancer = pd.read_csv("data/유방암.csv",engine = 'python')
df_breast_cancer.head()
```

```python
# 목표변수, 설명변수 역할 지정
df_breast_cancer_y = df_breast_cancer['diagnosis'] # 목표변수
df_breast_cancer_x = df_breast_cancer.drop('diagnosis', axis = 1, inplace = False)

print("목표변수 데이터 크기 : {}".format(df_breast_cancer_y.shape))
print("설명변수 데이터 크기 : {}".format(df_breast_cancer_x.shape))
```

```python
#절편추가
df_breast_cancer_x_const = sm.add_constant(df_breast_cancer_x)
df_breast_cancer_x_const.head()
```

```python
df_breast_cancer_x_const.describe()
```

```python
#회귀모델 생성
reg_multi_model = sm.OLS(df_breast_cancer_y,df_breast_cancer_x_const)
#적합
reg_multi_results = reg_multi_model.fit()
print(reg_multi_results.summary())
```

```python
#다중공정성확인
ds_vif = pd.DataFrame()
ds_vif["variable"] = df_breast_cancer_x_const.columns
ds_vif["VIF"] = [variance_inflation_factor(df_breast_cancer_x_const.values,i)for i in range(df_breast_cancer_x_const.shape[1])]
#VIF가 낮은 순으로 정렬
ds_vif.sort_values("VIF",inplace= True)
ds_vif.round(3)
```

```python
# STEP 5. 후진제거법을 이용한변수 선택
#RFE(Recursive feature elimination)함수:입력된 모델을 이용하여 변수중요도가 낮은 순으로 순차적으로 제거
#sklearn의 선형회귀 모델
model = LinearRegression()
#전체의 변수중 9개의 변수가 남을 때까지 변수 제거
# 변수를 10개 선택시 , concavity error의 pvalue 가 높게나온다, 9개로 설정해야 지워진다.
rfe = RFE(estimator = model, n_features_to_select=9).fit(df_breast_cancer_x,df_breast_cancer_y)
#선택된 변수
selected_cols = df_breast_cancer_x.columns[rfe.support_]
#제거된 변수
removed_cosl = df_breast_cancer_x.columns[~rfe.support_]

print("Selected Variables :{}".format(selected_cols))
print("Removed Variables :{}".format(removed_cosl))
```

```python
# 후진제거법을 이용한 변수선택 - 회귀모델 선택
df_breast_cancer_x_rfe = sm.add_constant(df_breast_cancer_x_const[selected_cols])
#회귀모델 생성
reg_multi_rfe_model = sm.OLS(df_breast_cancer_y,df_breast_cancer_x_rfe)
#적합
reg_multi_rfe_results = reg_multi_rfe_model.fit()
print(reg_multi_rfe_results.summary()) 
```

```python
# 후진제거법을 이용한 변수 선택 -선정된 변수들간의 다중공정성 확인
ds_rfe_vif = pd.DataFrame()
ds_rfe_vif["variable"] = df_breast_cancer_x_rfe.columns
ds_rfe_vif["VIF"] = [variance_inflation_factor(df_breast_cancer_x_rfe.values,i)for i in range(df_breast_cancer_x_rfe.shape[1])]
ds_rfe_vif.sort_values("VIF",inplace=True)
ds_rfe_vif.round(3)
```

```python
# STEP 6. 최종도출
df_reg_coef = pd.DataFrame({"Coef":reg_multi_rfe_results.params.values[1:]},index = reg_multi_rfe_model.exog_names[1:])
df_reg_coef.plot.barh(y = "Coef",legend = False)
```

```python
# 변수 중요성파악
# 표준화 회귀계수를 이용한 변수 중요도 확인
# 왜냐하면, 변수간에도 중요도 차이가 있다. - 표준화(scaling 필요)

df_breast_cancer_x_scaled = scale(df_breast_cancer_x_const[selected_cols])

df_breast_cancer_x_scaled = sm.add_constant(df_breast_cancer_x_scaled)

reg_multi_scaled_model = sm.OLS(df_breast_cancer_y,df_breast_cancer_x_scaled)

reg_multi_scaled_results = reg_multi_scaled_model.fit()

# 회귀계수의 방향은 동일하나 중요도의 크기가 변함 
df_reg_scaled_coef = pd.DataFrame({"Coef":reg_multi_scaled_results.params.values[1:]},
index = reg_multi_rfe_model.exog_names[1:])
df_reg_scaled_coef.plot.barh(y="Coef", legend = False)
```
