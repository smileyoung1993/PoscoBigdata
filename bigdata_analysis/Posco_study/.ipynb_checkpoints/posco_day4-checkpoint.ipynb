<h2>데이터 분석<h2>

```python
from statsmodels.tools.eval_measures import rmse
ds_repair = pd.read_csv("부품수리시간.csv",engine = 'python')
ds_repair
```

```python
ds_repair.plot.scatter(x = 'UNITS',y= 'MINUTES')
```

```python
ds_repair.corr(method = "pearson").round(3)
```

```python
# 회귀모델 생성
ds_repair_x = ds_repair['UNITS']
ds_repair_y = ds_repair["MINUTES"]

ds_repair_x_const = sm.add_constant(ds_repair_x) # 상수 coef 베타0,베타1을구함

reg_model = sm.OLS(ds_repair_y,ds_repair_x_const) #회귀모델에 대입
reg_result = reg_model.fit()# 회귀 fit 적용
print(reg_result.summary())# y_hat 회귀분석 요약
```

```python
# 등분산성/독립성검정
sns.set(style = 'whitegrid')
sns.residplot(ds_repair_x,ds_repair_y,lowess = True)
sns.jointplot(ds_repair_x,ds_repair_y)
# sns.regplot(ds_repair_x,ds_repair_y)
```

```python
# 정규화
obs = ds_repair_x.values
z= (obs-np.mean(obs))/np.std(obs) # 표준화
# sns.distplot(z, fit = norm)
# 정규확률도
stats.probplot(reg_result.resid, plot = plt)
```
