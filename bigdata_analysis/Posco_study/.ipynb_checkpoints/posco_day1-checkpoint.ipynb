```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns 

np.random.seed(seed = 1234)
normal = np.random.normal(20,2,100000)# 평균 20, 표준편차 2인 샘플 100000
sns.distplot(normal)
```

누적확률분포,정규분포

```python
norm = stats.norm(20,2) # 정규분포 정의 평균 = 20, 표준편차 2
xx = np.linspace(10,30,400) # x축정의
plt.plot(xx,norm.pdf(xx))
```

```python
norm = stats.norm(loc=20,scale=2) # 정규분포 정의 평균 = 20, 표준편차 2
xx = np.linspace(10,30,400) # x축정의 10~ 30 사이 개수는 400개
plt.plot(xx,norm.pdf(xx)) # proberity density function
```

```python
#cdp : 누적확률
#ppf : 역누적함수 - 확률을 주면 실제값을 준다
#rus : 랜덤변수 sampling
normal3 = stats.norm(loc = 115,scale = 20) # 정규분포 평균이 115 표준편차가 20
xx = np.linspace(60,170,400)
plt.plot(xx,normal3.pdf(xx))# 확률밀도
x = 135
v2 = stats.norm.cdf(x,115,20) # 누적확률 x = 135, 표준편차 20
print("{0}분이상 걸리는 as작업비율 : {1:.1f}%".format(x,(1-v2)*100))
v3 = stats.norm.sf(x,115,20) # 역누적함수 값으로 부터 +무한대의 누적값을 가져온다
# 1에서 빼줄 필요가 없다
print("{0}분이상 걸리는 as작업비율 : {1:.1f}%".format(x,v3*100))

inverse = stats.norm.ppf(v2,loc = 115, scale = 20)
inverse
```

```python
# 실습
# 평균 175, 표준편차 5

normal4 = stats.norm(loc = 175,scale = 5) # 정규분포 평균이 175 표준편차가 20
xx = np.linspace(160,190,400)
plt.plot(xx,normal4.pdf(xx))# 확률밀도
x = 180
v2 = stats.norm.cdf(x,175,5) # 누적확률 x = 175, 표준편차 20
v3 = stats.norm.sf(x,175,5) # sf
print("{0}cm 이상 경우 상위 : {1:.1f}%".format(x,(1-v2)*100))
print("{0}cm 이상 경우 상위 : {1:.1f}%".format(x,v3*100))
```

```python
#실습2
#평균 = 12, 표준편차 = 3
# 1리터를 가지고 12km이상, 15km이하를 달릴 확률
# 1리터를 가지고 9km이상, 15km이하를 달릴 확률
# 1리터를 가지고 15km이상을 달릴 확률
normal5 = stats.norm(12,3) # 정규분포 평균이 12 표준편차가 3
xx = np.linspace(3,21,400)
plt.plot(xx,normal5.pdf(xx))# 확률밀도
x1 = 12
x2 = 15
x3 = 9
v1 = stats.norm.cdf(x1,12,3) # 누적확률 x = 12, 표준편차 3
v2 = stats.norm.cdf(x2,12,3) # 누적확률 x = 15, 표준편차 3
v3 = stats.norm.cdf(x3,12,3)

print("{0}km 이상 {1}km 이하 경우 확률 : {2:.1f}%".format(x1,x2,(v2-v1)*100))
print("{0}km 이상 {1}km 이하 경우 확률 : {2:.1f}%".format(x1,x2,(v2-v3)*100))
print("{0}km 이상 경우 확률 : {1:.1f}%".format(x2,(1-v2)*100))
```

와이블분포

```python
np.random.seed(seed=1234)
xx = np.linspace(0,3000,400)

for c in [1,3,5,25,50]:
    weibull_df = stats.weibull_min(c=c,scale = 1200) # 내장method weibull_min
    plt.plot(xx,weibull_df.pdf(xx),label = c)
plt.legend()
```

```python
xx = np.linspace(0,3000,400)
# 그래프
c= 2.2 # 형상 모수
scale = 1200 # 척도 모수
weibull_df = stats.weibull_min(c=c,scale=scale) # 와이블 분포
plt.plot(xx,weibull_df.pdf(xx))
# 1500 시간이상일때의 확률
x= 1500
v4 = weibull_df.cdf(x) # 누적분포
print("P(x>={0}):{1:.4f}".format(x,1-v4))
v5= weibull_df.sf(x) # 누적분포
print("P(x>={0}):{1:.5f}".format(x,v5))
```

이항분포

```python
n=3
for i in range(n+1):
    # 이항분포 Probability Mass Function 
    prop = stats.binom.pmf(k=i,n=n,p=0.4) # 시행횟수(n),확률(p)
    print("P(x<={0}):{1:.3f}".format(i,prop))
```

```python
# 이항분포는 누적이라는 말이 없다.
#연습
n=5
for i in range(n+1):
    prop = stats.binom.pmf(k = i,n=n,p=0.3) # 확률이 0.3인 이항분포
    print("P(x<={0}):{1:.5f}".format(i,prop))
```

```python
for n in [3,30,40,60]:
    binom_df = stats.binom(n=n,p=0.4) # p=0.4인 이항분포(정규)
    xx = np.linspace(0,n,n+1) # 0부터 n사이 n+1개의
    plt.plot(xx,binom_df.pmf(xx),label = n) #그래프 그리기(pmf)- 확률질량함수
plt.legend()
```

포아송분포

```python
# 평균
mu = 2
# 1분당 3번의 전화가 걸려올 확률
prop=stats.poisson.pmf(3,mu) # 확률값은 확률질량 pmf
# 1분당 2회 이하의 전화가 걸려올 확률 - 누적 cdf
prop_cdf= stats.poisson.cdf(2,mu)
print("P(x=={0}):{1:.5f}".format(3,prop))
print("P(x<={0}):{1:.5f}".format(2,prop_cdf))
```

```python
#포아송 실습
mu = 1
prop = stats.poisson.pmf(2,mu) # 2명의 손님이 도착할 확률
#1명 이하의 손님이 도착할 확률 - 누적
prop_cdf = stats.poisson.cdf(1,mu)
# 3명 이상의 손님이 도착할 확률
prop_sf= stats.poisson.sf(3,mu)
print("P(x=={0}):{1:.5f}".format(2,prop))
print("P(x<={0}):{1:.5f}".format(1,prop_cdf))
print("P(x>={0}):{1:.5f}".format(3,prop_sf))
```

```python
for mu in [2,4,6,10]: # 평균이 커질수록 정규분포에 가까워진다.
    poisson_df = stats.poisson(mu = mu)
    xx = np.linspace(0,3*mu,3*mu+1)
    plt.plot(xx,poisson_df.pmf(xx),label= mu)
v1= poisson_df.pmf(2)
print("P(x =={0}):{1}".format(2,v1))
plt.legend()
```

```python
poisson_df1 = stats.poisson(mu=2) # 평균이 2인 포아송분포 정의
v_p1 = poisson_df1.pmf(3) #확률질량함수
print("{0}:확률질량함수".format(v_p1))
v_p2 = poisson_df1.cdf(2) #확률밀도함수
print("{0}:확률밀도함수".format(v_p2))
```

```python
count = df['Flaws'].value_counts().sort_index() # Flaws 변수 count:value_count, 순서대로 정렬 sort_index()
count
```

```python
#CumCnt 계산 : cumsum
cumsum = np.cumsum(count) # 누적합 계산
cumsum
```

```python
percent = count / sum(count)*100
percent
```

```python
cumpct = np.cumsum(percent)
cumpct
```

```python
#DataFrame 으로 취합
count_data =pd.DataFrame({'Count':count,'CumCnt':cumsum,'Percent':percent,'CumPct':cumpct})
count_data
```

t-test

```python
df = pd.DataFrame({"sample":[54.1,53.3,56.1,55.7,54.0,54.1,54.5,57.1,55.2,53.8,54.1,54.1,
                            56.1,55.0,55.9,56.0,54.9,54.3,53.9,55.0]})
lower,upper = stats.t.interval(0.95,len(df)-1,loc=np.mean(df),scale = stats.sem(df))
# 자유도 len(df)-1 , 0.95 : 신뢰도, loc = 평균, scale : 표준오차
print("신뢰구간:[{0},{1}]".format(lower.round(2),upper.round(2)))
```

```python
#실습 - 모표준편차를 모르는 경우
df = pd.DataFrame({"sample":[73,71,74,69,70,73,70,73,70,68,75,72,70,72,73,70,70,72,71,70,75,72]})
lower,upper = stats.t.interval(0.95,len(df)-1,loc=np.mean(df),scale = stats.sem(df))
print("신뢰구간:[{0},{1}]".format(lower.round(2),upper.round(2)))
```

ttest-1sample t

```python
df = pd.DataFrame({'sample':[74.5,81.2,73.8,82.0,76.3,75.7,80.2,72.6,77.9,82.8]})
# print(np.mean(df))
t_test = stats.ttest_1samp(df,76.7)
t,p = t_test.statistic.round(3), t_test.pvalue.round(3)
print('1-sample t-test')
print("t:{}".format(t))
print("p:{}".format(p))
```

```python
# 실습 - ttest - 1 sample t
df = pd.DataFrame({'data':[85.0,79.0,79.1,79.9,81.6,78.6,85.4,83.4,78.1,79.2]})
t_test = stats.ttest_1samp(df,80.93)
t,p = t_test.statistic.round(3), t_test.pvalue.round(3)
print('1-sample t-test')
print("t:{}".format(t))
print("p:{}".format(p))
#print(개선활동이 만족도를 변화시키지 않았다. 귀무가서 채택)
```

2-sample t

```python
df1 = pd.DataFrame({"sample": [6,5,5,4,6,7,6,4,5,6,4,5,5,6,4,8,6,5,6,7]})
df2 = pd.DataFrame({"sample": [7,5,7,8,7,8,8,5,7,6,5,5,6,6,5,7,9,7,7,8]})
t_test = stats.ttest_ind(df1,df2) # individual = ind
t,p = t_test.statistic.round(3),t_test.pvalue.round(3)
print("2-sample t-test")
print("t:{}".format(t))
print("p:{}".format(p))
```

```python
#실습 2-sample t
df_A = pd.DataFrame({"sample":[1.883,1.715,1.799,1.768,1.711,1.832,1.427,1.344]})
df_B = pd.DataFrame({"sample":[1.435,1.572,1.486,1.511,1.457,1.548,1.404,1.883]})
# 2 - sample test
t_result = stats.ttest_ind(df_A,df_B)
# 2 - sample test t,p값 저장
t,p = t_result.statistic.round(3),t_result.pvalue.round(3)
# 출력
print("2-sample t-test")
print("t:{}".format(t))
print("p:{}".format(p))
```

paired -T test

```python
#실습 paired  -T test
df_tb = pd.DataFrame({"test_before":[720,589,780,648,720,589,780,648,780,648]})
df_ta = pd.DataFrame({"test_after":[710,580,787,712,750,600,782,670,790,680]})

t_result = stats.ttest_rel(df_tb,df_ta) # paired t

t,p = t_result.statistic.round(4),t_result.pvalue.round(4)
print("Paired t-test")
print("t검정통계량:{}".format(t))
print("p-value:{}".format(p))
print("p값이 0.0497이므로 학원수강 전후의 점수는 차이가 있다고 할 수 있다")
```

아노바 - oneway

```python
df1 = pd.DataFrame({"A":[892,623,721,678,723,790,720,670,690,771]})
df2 = pd.DataFrame({"B":[672,743,761,628,823,720,750,870,790,721]})
df3 = pd.DataFrame({"C":[893,627,726,673,713,789,722,660,670,761]})
f_test = stats.f_oneway(df1,df2,df3)
f,p = f_test.statistic.round(3),f_test.pvalue.round(3)
print("one way ANOVA")
print("F:{}".format(f))
print("P:{}".format(p))
```

1 - Proportion test

```python
# 변수 선언
count = 15
nobs = 100
value = .1
# proportion test 실행
stat, pval = proportions_ztest(count,nobs,value)
# 결과 출력
print("1 Proportion test")
print("P 검정통계량 : {0:0.3f}".format(stat))
print("P-value : {0:0.3f}".format(pval))
print("P-value가 0.161 > 0.05 이므로 품질개선 결과 기존 대비 만족도가 차이가 없다")
```

```python
# 1 Proportion test 실습
count = 10
nobs =100
value =.2
# proportion test 실행
stat, pval = proportions_ztest(count,nobs,value)
# 결과 출력
print("1 Proportion test")
print("P 검정통계량 : {0:0.3f}".format(stat))
print("P-value : {0:0.3f}".format(pval))
print("P-value가 0.001 < 0.05 이므로 품질개선 결과 기존 대비 만족도가 차이가 있다")
```

2 propotion test

```python
# 변수 선언
count = np.array([4,1])
nobs = np.array([1000,1200])

# proportion test 실행
stat, pval = proportions_ztest(count,nobs)

# 결과 출력
print("2 Proportion test")
print("P 검정통계량 : {0:0.3f}".format(stat))
print("P-value : {0:0.3f}".format(pval))
print("P-value가 0.120 > 0.05 이므로 두 공정의 불량률이 차이가 있다고 할 수 없다.")
```

```python
# 2 proportion test 실습
count = np.array([14,5])
nobs = np.array([1200,1200])

# proportion 실행
stat,pval = proportions_ztest(count,nobs)

# 결과 출력
print("2 Proportion test")
print("P 검정통계량 : {0:0.3f}".format(stat))
print("P-value : {0:0.3f}".format(pval))
print("P-value가 0.038 < 0.05 이므로 두 공정의 불량률이 차이가 있다고 할 수 있다.")
```

카이제곱 검정

```python
#df 생성
df = pd.DataFrame({'HSG':[270,260,236,234],'SS':[228,285,225,262],'SPA':[277,284,231,208]})
# Chi-square test 실행
# 카이제곱값, P-value, 자유도,기대값
chi,pval,dof,expected = stats.chi2_contingency(df.T)
# 결과 출력
print("chi-square test")
print("chisq : {0:0.3f}".format(chi))
print("p:{0:0.3f}".format(pval))
print("degree of freedom:{}".format(dof))
print("expected value : \n{}".format(expected.round(3)))
```

```python
#카이제곱 실습
df = pd.DataFrame({'AS':[18,8,4,4,3,3],'PRICE':[1,2,1,1,1,25],'SPEC':[8,14,3,2,3,8],'EXPEND':[7,5,4,3,1,10],'DESIGN':[10,5,9,2,1,2],'SAFETY':[9,9,5,7,1,1],'FUNC':[10,4,4,3,1,7]})
# Chi-square test 실행
# 카이제곱값, P-value, 자유도,기대값
chi,pval,dof,expected = stats.chi2_contingency(df.T)
# 결과 출력
print("chi-square test")
print("chisq : {0:0.3f}".format(chi))
print("p:{0:0.3f}".format(pval))
print("degree of freedom:{}".format(dof))
print("expected value : \n{}".format(expected.round(3)))
print("P-value : 0.000 < 0.05 이므로 각 제품에 대해 이미지에 대한 차이가 있다고 할 수 있다.")
```
