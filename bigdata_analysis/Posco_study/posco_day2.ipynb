```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns 
import statsmodels.api as sm
# from statsmodels.stats.proportion import proportions_ztest
```

상관분석

```python
# 연습
df1 = [1,2,3,4,4,5,6,6,7,8]
df2 = [23,29,49,64,74,87,96,97,109,119]
plt.scatter(df1,df2)
# Correlation Analysis 실행
corr,pval =stats.pearsonr(df1, df2)
print("correlation analysis")
print("corr: {0:0.3f}".format(corr))
print("p:{0:0.3f}".format(pval))
```

```python
#실습
df1 = [37.6,38.6,37.2,36.4,38.6,39,37.2,36.1,35.9,37.1,36.9,37.5,36.3,38.1,39,36.9,36.8,37.6,33,33.5]
df2 = [14,15,14,16,17,14,17,16,15,16,14,16,18,13,15,17,14,16,20,21]
plt.scatter(df1,df2)

corr,pval = stats.pearsonr(df1,df2) # 두 변수의 상관도, Pval를 구한다.
print("correlation analysis")
print("corr: {0:0.3f}".format(corr))
print("p: {0:0.3f}".format(pval))
# pvalue는 유의수준에 따라, 해석된다.
# 만약 유의수준이 5%이면, 95%이상의 확률이 나와야 상관성이 있다고 보므로 pval이 0.05보다 작아야 상관성이 있다고 본다.
```

회귀분석(Simple Linear Regression)

```python
#연습
minutes =[1,2,3,4,4,5,6,6,7,8]
units = [23,29,49,64,74,87,96,97,109,119]
# 두 변수의 산점도를 확인한다.
plt.scatter(minutes,units)

#simple linear regression 실행
minutes=sm.add_constant(minutes)
model = sm.OLS(units,minutes)
result = model.fit()
# simple linear regression 결과 출력
result.summary()
print(result.summary())
print("부품수와 부품수리간의 상관 R-squared는 0.979이므로 상관성이 매우 높다고 본다.")
```

```python
#실습
strength = [37.6,38.6,37.2,36.4,38.6,39,37.2,36.1,35.9,37.1,36.9,37.5,36.3,38.1,39,36.9,36.8,37.6,33,33.5]
bake_time= [0.798922,0.84902,0.816163,0.758266,0.815894,0.807477,0.809068,0.7664,0.742243,0.751118,0.740629,0.751495,
            0.738863,0.827428,0.818172,0.740939,0.757128,0.804695,0.695121,0.735377]

plt.scatter(strength,bake_time)
strength = sm.add_constant(strength)
model = sm.OLS(bake_time,strength)
result = model.fit()
result.summary()
print(result.summary())
print("초코칩의 강도와 굽는시간의 상관 R-squared는 0.646이므로 상관성이 있다고 본다.")
```

```python
# 데이터 변환 연습 (pivot,melt)
df = pd.DataFrame({'foo':['one','one','one','two','two','two'],
                  'bar':['A','A','C','A','B','C'],
                  'baz':[1,2,3,4,5,6],
                  'zoo':['x','y','z','q','w','t']})
df
# bar에서 'A''A'가 두개이므로 Pivot이 되지 않는다. but pivot_table은 가능
```

```python
df.pivot(index = 'foo',columns = 'bar', values = 'baz')
df.pivot(index = 'foo', columns = 'bar')['baz']
```

```python
df.pivot_table(index = 'foo',columns = 'bar', values = 'baz',aggfunc = sum)
# 원소를 맞추지 않아도 pivot_table은 만들어준다.
```
