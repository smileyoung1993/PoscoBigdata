```python
#데이터 표준화
scaler = StandardScaler()
#데이터표준화변환
df_flower_x_scaled = scaler.fit_transform(df_flower_x)
#표준화된 데이터 확인
print("Scaled Data :\n{}".format(df_flower_x_scaled[:5].round(3)))
print("Shape :\n{}".format(df_flower_x_scaled.shape))
```

```python
# 주성분 분석모델 생성 Prin1 ~ 6 까지
pca = PCA(n_components=4)
x_pca = pca.fit_transform(df_flower_x_scaled)
#계산된 Prin1~6까지의 Row개수, column개수 화인
print("PCA DATA : \n{}".format(x_pca[:3]))
#shape : Prin 1~6까지 Row개수, column개수 확인
print("Shape: \n{}".format(x_pca.shape))
#주성분 점수 저장
df_x_pca = pd.DataFrame(x_pca,columns = ["Prin1","Prin2","Prin3","Prin4"])
```

```python
# 고유값 확인
#주성분 통계량 입력용 새로운 데이터 프레임 생성
df_pca = pd.DataFrame()
#주성분 분석 모델의 주성분 번호, Elgenvalue, Eigenvalue Ratio 등을 데이터 프레임의 형태로 저장
df_pca["Prin"] = [i for i in range(1,5)]
df_pca["EigenValues"] = pca.explained_variance_##Eigen values
df_pca["EigenValueRatio"] = pca.explained_variance_ratio_## Eigenvalue Ratio
df_pca["CumsumEigenValueRatio"] = np.cumsum(pca.explained_variance_ratio_)
df_pca.round(3)
```

```python
# 주성분 통계량 입력용 새로운 데이터 프레임 생성
df_feature = pd.DataFrame({"Feature":df_flower_x.columns})
# 주성분 prin1~6까지의 고유벡터 저장
df_eigenvectors = pd.concat([df_feature,pd.DataFrame(pca.components_,columns = ["Prin1","Prin2","Prin3","Prin4"])],axis =1)
df_eigenvectors
```

```python
# 고유값 비율 그래프 생성
plt.plot(df_pca["Prin"],df_pca["EigenValueRatio"],label ="Ratio")
# 누적된 고유값 비율 그래프 생성
plt.plot(df_pca["Prin"],df_pca["CumsumEigenValueRatio"],label = "Cumsum Ratio")
# 그래프 label, lengend 설정
plt.ylabel("Eigen Value ratio")
plt.xlabel("Prin")
plt.legend(loc = "best")
```

```python
#회귀분석 패키지 불러오기
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
#Scree plot 을 이용하여 선택한 주성분
x_var = ["Prin1","Prin3"]
# 기존 주성분 데이터에 절편 컬럼 추가
df_x_pca_const = sm.add_constant(df_x_pca[x_var])
#train test 분할
# random_state 에 따라, 다르게 분류하므로 당연히, 설명력이 달라질 수 있다.
x_tr,x_test,y_tr,y_test = train_test_split(df_x_pca_const,df_flower_y,test_size =0.3 , random_state = 2222)

#회귀모델 생성
reg_model = sm.OLS(y_tr,x_tr)
reg_model = reg_model.fit()
print(reg_model.summary())
```
