```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import scale

import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib

for f in fm.fontManager.ttflist:
    if 'NanumBarunGothic' in f.fname:
        plt.rcParams['font.family'] = 'NanumBarunGothic' # 폰트를 나눔고딕으로
print(plt.rcParams['font.family'])
matplotlib.rcParams['axes.unicode_minus'] = False #'-'기호 제대로 나오게함
```

```python
df_hmeq = pd.read_csv("../data/HMEQ.csv",encoding = 'euc-kr')
df_hmeq.head()
```

```python
df_hmeq.fillna(df_hmeq.mean(),inplace=True)
```

```python
# 숫자형 데이터 만을 불러온다.
df_numeric = df_hmeq.select_dtypes(["int64","float64"])
df_numeric.head()
```

```python
df_numeric_y = df_numeric["BAD"]
```

```python
df_numeric_y
```

```python
df_numeric_x = df_numeric.drop("BAD",axis = 1)
```

```python
df_numeric_x
```

```python
#scaled 
df_numeric_x_scale = scale(df_numeric_x,axis = 0)
```

```python
df_numeric_x_scale = pd.DataFrame(df_numeric_x_scale, columns=df_numeric_x.columns)
```

```python
df_numeric_x_scale
```

```python
np.random.seed(seed = 1234)
x= df_numeric_x_scale
y= df_numeric_y

# data를 split 하기전에 scale하는 것이 맞다.
x_tr,x_test,y_tr,y_test = train_test_split(x,y,test_size = 0.3,random_state = 1234)

print("train data x size : {}".format(x_tr.shape))
print("train data y size : {}".format(y_tr.shape))
print("test data x size : {}".format(x_test.shape))
print("test data y size : {}".format(y_test.shape))
```

```python
# 가우시안 모델 생성
naive_model = GaussianNB()
# 데이터 적합
naive_model.fit(X=x_tr,y=y_tr)
# 사전확률
naive_model.class_prior_.round(3)
```

```python
print("Accuracy : {0:.3f}".format(naive_model.score(x_test,y_test)))
```
