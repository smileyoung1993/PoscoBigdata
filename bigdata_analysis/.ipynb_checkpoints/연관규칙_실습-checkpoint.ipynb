```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df_product = pd.read_csv("./data/상품구매.csv",engine = "python")
df_product
```

```python
df_p=pd.melt(df_product, id_vars='ID')
```

```python
df_product.columns
```

```python
df_product.ID.value_counts()
```

```python
df_product.PRODUCT.value_counts()
```

```python
df_product.groupby("ID").agg(lambda x : list(x))
```

```python
# 시리지
df_product.groupby("ID").agg(lambda x:list(x)).PRODUCT
```

```python
# 리스트로 변환
df_product.groupby("ID").agg(lambda x:list(x)).PRODUCT.tolist()
```

```python
# ID로 묶고, 리스트만든후(집계함수이용)-> PRODUCT만을 뽑아내면(시리즈)->tolist로 리스트로 다시변환  
df_p = df_product.groupby("ID").agg(lambda x:list(x)).PRODUCT.tolist()
```

```python
# items 중 index 0번이 갖고 있으면 True, 아니면 False
enc = TransactionEncoder()
arr = enc.fit_transform(X=df_p)
# 위의 함수는 데이터를 행렬의 형태로 반환 -> pandas의 DataFrame의 형태로 변환
df_pre = pd.DataFrame(arr,columns=enc.columns_)
df_pre
```

```python
# 최소 지지도
min_support = 0.3 # 2/4
# 지지도 계산
df_freq = apriori(df_pre,min_support=0.3,use_colnames=True)
df_freq
```

```python
#기준 = 신뢰도, 최소 신뢰도 = 0.5
df_asso_rule = association_rules(df_freq,metric="confidence",min_threshold= 0.5)
df_asso_rule.round(3)
```

<h3> 한번에 데이터 전처리 하기 <h3>
    - pivot, isnull(), ~ 이용하기

```python
df_product1 =df_product
df_product1
```

```python
# null 인건 true, 아닌건 false 이므로 이를 inverse시키면 가능합니다.
df_product1 = ~df_product1.pivot(index = "ID", columns= 'PRODUCT', values= 'PRODUCT').isnull()
df_product1
```

```python
# 최소 지지도
min_support = 0.3 # 2/4
# 지지도 계산
df_freq = apriori(df_product1,min_support=0.3,use_colnames=True)
df_freq.round(3)
```

```python
#기준 = 신뢰도, 최소 신뢰도 = 0.5
df_asso_rule = association_rules(df_freq,metric="confidence",min_threshold= 0.5)
df_asso_rule.round(3)
```
