```python
from sklearn.decomposition import PCA
from scipy import stats
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import scale
from scipy.cluster.hierarchy import linkage, dendrogram,fcluster
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib
from sklearn.cluster import KMeans,DBSCAN

for f in fm.fontManager.ttflist:
    if 'NanumBarunGothic' in f.fname:
        plt.rcParams['font.family'] = 'NanumBarunGothic' # 폰트를 나눔고딕으로
print(plt.rcParams['font.family'])
matplotlib.rcParams['axes.unicode_minus'] = False #'-'기호 제대로 나오게함
```

```python
df_flower = pd.read_csv("./data/IRIS.csv",encoding = 'euc-kr')
df_flower.head()
```

```python
# 목표변수가 범주형 string 값을 가지므로 factorize 함수를 사용해서, 숫자로 mapping 해준다.
df_flower.SPECIES = pd.factorize(df_flower.SPECIES)[0]
df_flower
```

```python
pd.factorize(df_flower.SPECIES)
# 여기서 pd.factorize(df_flower.SPECIES)[0]을해주면 숫자로만 이루어진 튜플을 가져올 수 있다.
```

```python
#데이터 표준화
scaler = StandardScaler()
#데이터표준화변환습
df_flower_scaled = scaler.fit_transform(df_flower)
#표준화된 데이터 확인
print("Scaled Data :\n{}".format(df_flower_scaled[:5]))
print("Shape :\n{}".format(df_flower_scaled.shape))
```

```python
# 목표변수
df_flower_y = df_flower_scaled[:,0]
# 설명변수
df_flower_x = df_flower_scaled[:,1::]
df_flower_x
```

```python
# 연결 ward,거리계산:Euclidean
Z_ward = linkage(df_flower_x,method="ward",metric="euclidean")
# 연결:average, 거리계산 : Euclidean
Z_average = linkage(df_flower_x,method="average",metric = "euclidean")
# 연결:average, 거리계산 : Minkowski
Z_Minkowski = linkage(df_flower_x,method="average",metric="minkowski")
# 덴드로그램[연결:ward, 거리계산: Euclidean]
plt.figure(figsize = (10,10))
ax = plt.subplot(111)
dendrogram(Z_ward,leaf_font_size=10,orientation="right")
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("distance")
plt.ylabel("sample index")
```

```python
# 덴드로그램[average, 거리계산: Euclidean]
plt.figure(figsize = (10,10))
ax = plt.subplot(111)
dendrogram(Z_average,leaf_font_size=10,orientation="right")
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("distance")
plt.ylabel("sample index")
```

```python
# 덴드로그램[연결:Minkowski, 거리계산: Euclidean]
plt.figure(figsize = (10,10))
ax = plt.subplot(111)
dendrogram(Z_Minkowski,leaf_font_size=10,orientation="right")
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("distance")
plt.ylabel("sample index")
```

```python
# linkage 함수를 통해 얻은 Z_ward 행렬 데이터를 이용하여 군집 생성
cluster_hierarchical = fcluster(Z=Z_ward, t =5, criterion= "distance")
# 데이터에 계층형 군집 결과 입력
df_flower["ClusterHierarchical"] = cluster_hierarchical
df_flower
```

```python
# 군집별 데이터 요약
df_summary_hc = df_flower.groupby("ClusterHierarchical").agg(["min","mean","max"])
# 집계 함수를 이용할 경우 열 이름이 두 줄에 걸쳐 나오므로 한 줄로 변환
level0 = df_summary_hc.columns.get_level_values(0)
level1 = df_summary_hc.columns.get_level_values(1)
df_summary_hc.columns = level0 + "_" + level1
df_summary_hc.reset_index(inplace = True)
df_summary_hc.round(3)
```

```python
# 최대군집 생성개수
n_iter_cluster =15
cluster_range = [i+1 for i in range(n_iter_cluster)]
clus_err = []
for n_clus in cluster_range:
    clus = KMeans(n_clus)
    clus.fit(df_flower_x)
# 각 데이터로부터 가장 가까운 군집 중심점까지 거리 제곱합
    clus_err.append(clus.inertia_)
df_error = pd.DataFrame({"NumberofCluster":cluster_range,"Error":clus_err})
df_error.round(3)
```

```python
# K 평균 군집 분석
# 군집별 군집 중시점까지 거리 제곱합을 시각화
plt.figure(figsize= (10,10))
plt.plot(df_error["NumberofCluster"],df_error["Error"])
plt.title("Sum of squared distance")
plt.xlabel("Clusters")
plt.ylabel("Sum of squared distance")
```

```python
# k-평균 군집 번호 생성
clus = KMeans(5)
clus.fit(df_flower_x)
cluster_kmeans = [i+1 for i in clus.labels_]
df_flower["ClusterKmeans"] = cluster_kmeans
df_flower.head()
```

```python
df_summary_km = df_flower.drop("ClusterHierarchical",axis =1,inplace = False).\
groupby("ClusterKmeans").agg(["min","max","mean"])

level0 = df_summary_km.columns.get_level_values(0)
level1 = df_summary_km.columns.get_level_values(1)

df_summary_km.columns = level0 + "_" + level1
df_summary_km.reset_index(inplace = True)
df_summary_km.round(3)
# 각 군집별로 변수들이 다른 통계량 값들을 갖고 있음
```

```python
# eps : 특정 데이터로부터 다른 데이터를 포함하는 거리 min_samples : 군집이 되기 위한 최소 데이터 수
dbscan = DBSCAN(eps = 1.4, min_samples =2).fit(df_flower_x)
# 군집결과
cluster_dbscan = dbscan.labels_
# sklearn의 DBSCAN에서는 noise 데이터의 경우 군집이 -1로 표시되면 0번 군집부터 시작
# 다른계층형, K-평균은 1번 군집부터 시작하므로 다른 군집 번호와 통일.Noise 데이터는 0번 군집으로 표기
for(i,val) in enumerate(cluster_dbscan):
    if val < 0:
        cluster_dbscan[i] = 0
    else:
        cluster_dbscan[i] = cluster_dbscan[i] + 1
df_flower["ClusterDBScan"] = cluster_dbscan
df_flower.head()
```

```python
# DB 스캔 군집별 요약 통계량
df_summary_db = df_flower.drop(["ClusterHierarchical","ClusterKmeans"],axis = 1,inplace = False).\
groupby("ClusterDBScan").agg(["min","max","mean"])
level0 = df_summary_db.columns.get_level_values(0)
level1 = df_summary_db.columns.get_level_values(1)

df_summary_db.columns = level0 + "_" + level1 # 변수명 + 통계량
df_summary_db.reset_index(inplace = True)
df_summary_db.round(3)

# 각 군집별 변수들이 다른 통계량 값들을 갖고 있음
# cluserDVScan 군집 수만큼 테이블의 row가 생긴다.
```
