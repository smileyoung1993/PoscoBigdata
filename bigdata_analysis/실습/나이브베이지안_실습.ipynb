```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import scale

import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib

for f in fm.fontManager.ttflist:
    if 'NanumBarunGothic' in f.fname:
        plt.rcParams['font.family'] = 'NanumBarunGothic' # 폰트를 나눔고딕으로
print(plt.rcParams['font.family'])
matplotlib.rcParams['axes.unicode_minus'] = False #'-'기호 제대로 나오게함
```

```python
df_custom = pd.read_csv("../data/callcustom.csv",encoding = 'euc-kr')
df_custom.head()
```

```python
df_custom.CHURN = df_custom.CHURN.map(lambda x:1 if x =="Active" else 0)
df_custom
```

```python
df_numeric = df_custom.select_dtypes(["int64","float64"])
df_numeric.head()
```

```python
df_numeric_x = df_numeric.drop("CHURN",axis =1)
df_numeric_y = df_numeric["CHURN"]
```

```python
np.random.seed(seed = 1234)
x= df_numeric_x
y= df_numeric_y

# data를 split 하기전에 scale하는 것이 맞다.
x_tr,x_test,y_tr,y_test = train_test_split(x,y,test_size = 0.3,random_state = 1234)

print("train data x size : {}".format(xhttps://github.com/smileyoung1993/Posco_bigdata.git_tr.shape))
print("train data y size : {}".format(y_tr.shape))
print("test data x size : {}".format(x_test.shape))
print("test data y size : {}".format(y_test.shape))
```

```python
# 가우시안 모델 생성
naive_model = GaussianNB()
# 데이터 적합
naive_model.fit(X=x_tr,y=y_tr)
# 사전확률
naive_model.class_prior_.round(3)
```

```python
print("Accuracy : {0:.3f}".format(naive_model.score(x_test,y_test)))
```
