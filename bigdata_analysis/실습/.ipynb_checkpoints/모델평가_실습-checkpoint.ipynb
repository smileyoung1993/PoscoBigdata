```python
# 필요한 라이브러리 import
import pandas as pd
import numpy as np
from sklearn.preprocessing import scale
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib
import time
for f in fm.fontManager.ttflist:
    if 'NanumBarunGothic' in f.fname:
        plt.rcParams['font.family'] = 'NanumBarunGothic' # 폰트를 나눔고딕으로
print(plt.rcParams['font.family'])
matplotlib.rcParams['axes.unicode_minus'] = False #'-'기호 제대로 나오게함
```

```python
# 데이터 불러오기
df_custom = pd.read_csv("../data/callcustom.csv",engine = 'python')
df_custom.head()
```

<h3>목표변수 이산형으로 변형<h3>

```python
df_custom.CHURN = df_custom.CHURN.map(lambda x:1 if x =="Active" else 0)
df_custom
```

<h3> 불필요한 데이터 삭제 <h3>

```python
df_custom.drop("CUSTOMER_ID",axis=1,inplace = True)
df_custom.columns
```

<h3> Train,Test데이터 분리 <h3>

```python
df_custom_y = df_custom["CHURN"]
df_custom_x = pd.get_dummies(df_custom)
df_custom_x.drop("CHURN",axis= 1,inplace = True)
```

```python
x= df_custom_x
y= df_custom_y
x_tr,x_test,y_tr,y_test = train_test_split(x,y,test_size = 0.3,random_state = 1234)

print("train data x size : {}".format(x_tr.shape))
print("train data y size df_eval.plot.bar(rot=0): {}".format(y_tr.shape))
print("test data x size : {}".format(x_test.shape))
print("test data y s토ize : {}".format(y_test.shape))
```

<h3> 데이터 스케일링 <h3>

```python
x_tr_scaled = scale(x_tr,axis = 0)
x_test_scaled = scale(x_test,axis = 0)
```

```python
# 정확도 저장 리스트
train_accuracy = []
test_accuracy = []
#f1 score 저장 리스트
model_f1_score = []
```

<h3> 반복루프를 통해 모델별 정확도 및 성능 저장 <h3>

```python
# loop
names = ["Rf","Gb","Dt","Knn","svm"]
clfs= [
    RandomForestClassifier(n_estimators= 500,max_depth=4,min_samples_leaf=30,criterion='gini',random_state=1234),
    GradientBoostingClassifier(learning_rate= 0.1,max_depth=4,min_samples_leaf=30,n_estimators=5,random_state=1234),
    DecisionTreeClassifier(max_depth=4,min_samples_leaf=30,criterion='gini'),
    KNeighborsClassifier(),
    SVC(C=6,gamma=0.1,random_state=1234)
]
plt.figure()
start = time.time()
for name,clf in zip(names,clfs):
    clf.fit(x_tr_scaled,y_tr)
    train_accuracy.append(clf.score(x_tr_scaled,y_tr))
    test_accuracy.append(clf.score(x_test_scaled,y_test))
#   y_pred = clf.predict(x_test_scaled)[:,1]
    model_f1_score.append(f1_score(y_test,clf.predict(x_test_scaled)))

```

<h3> 결과 테이블 <h3>

```python
df_eval = pd.DataFrame(index= names)
df_eval["TrainAccuracy"] = train_accuracy
df_eval["TestAccuracy"] = test_accuracy
df_eval["F1Score"] = model_f1_score
df_eval.round(3)
```

<h3> 모델별 성능 비교 그래프 <h3>

```python
df_eval.plot.bar(rot=0)
```
